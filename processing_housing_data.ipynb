```python
import os
import json
import pprint

```

```python
jsons = [x for x in os.listdir('C:\\Users\\chris\\Nextcloud\\Desktop') if '.json' in x]

json_data = []
for j in jsons:
    with open(j, 'r') as fp:
        json_data.append(json.load(fp))
```

```python
keys = ['addressDisplay', 'addressDisplayTruncated', 'addressUrl', 'bathroomsFull', 'bathroomsHalf', 'bathsDisplay', 'bedroomsTotal', 'bedsDisplay', 'buildingAreaTotal', 'city', 'closeDate', 'coListAgentFullName', 'countyOrParish', 'daysOnSite', 'garageSpaces', 'latitude', 'listAgentEmail', 'listAgentFullName', 'listAgentPreferredPhone', 'listAgentStateLicense', 'listAgentKeyNumeric', 'listTeamAgentDREs', 'listAgentMLS', 'listAgentProfilePublicYN', 'listingId', 'listOfficeName', 'price', 'priceFormatted', 'longitude', 'lotSizeArea', 'lotSizeAcres', 'mediaThumbURL', 'mediaURL', 'mlsStatus', 'modificationTimestamp', 'newListing', 'originatingSystemName', 'sourceSystemName', 'openHouseYN', 'onMarketDate', 'offMarketDays', 'propertyType', 'propertySubType', 'postalCode', 'photosCount', 'listingKeyNumeric', 'stateOrProvince', 'statusChangeTimestamp', 'standardStatus', 'standardStatusDisplay', 'streetDirPrefix', 'streetName', 'streetSuffix', 'transactionType', 'unparsedAddress', 'yearBuilt', 'virtualTourTrueFalse']
pprint.pprint(keys)
```

```python
filters = [('city', ['San Jose', 'Campbell', 'Cupertino', 'Santa Clara'])]

def check_filter(value, expected):
    #print(value, expected)
    if isinstance(expected, list):
        if value not in expected:
            return False
    if isinstance(expected, str):
        if expected not in value:
            return False
    return True

quick_set = set()
passed_inspection = {}
for listings in json_data:
    for listing in listings['listings']:
        cont = False
        #print(listing.get('propertySubType'), listing.get('city'))
        for filter in filters:
            filter_key, filter_value = filter
            if not check_filter(listing[filter_key], filter_value):
                cont = True
                break
        if cont:
            continue
        quick_set.add(listing.get('city'))
        passed_inspection[listing.get('listingId')] = listing

```

```python
pprint.pprint(passed_inspection)
print(quick_set)
```
